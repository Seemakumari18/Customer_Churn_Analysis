```python
## TENURE GROUP BANAO
```

```python
df['tenure_group'] = pd.cut(df['Tenure'],
                           bins=[0,12,24,60],
                           labels=['0-1 yr','1-2 yr','2+ yr'])
df['tenure_group']
```

```python
## RISK SEGMENT..
```

```python
df['risk_flag'] = ((df['Monthly_Charges'] > df['Monthly_Charges'].mean()) & 
                   (df['Tenure'] < df['Tenure'].mean())).astype(int)
df['risk_flag'] 
```

```python
## IMPORTANT GRAPH
## CHURN VS TENURE GROUP
```

```python
sns.countplot(x='tenure_group', hue='Churn', data=df)
plt.show() 
```

```python
"Customer with shorter tenure show higher tenure rates, suggesting 
early_stage dissatisfaction and highlighting the need for improved 
onboarding strategies."
```

```python
##Insight
“Customers with lower tenure (0–1 year) have the highest churn rate.”
“Churn decreases as customer tenure increases.”
“Long-term customers (2+ years) are more loyal and less likely to churn.”
## PROBLEMS
“New customers are leaving early before forming long-term relationships.”
“Onboarding or initial experience may not be strong enough.”
“Company is unable to retain customers in the early stage.”
## SOLUTIONS
“Improve onboarding experience for new customers.”
“Engage customers in the first few months with offers or support.”
“Provide welcome benefits or discounts for new users.”
“Build early trust through better service and communication.”
```
